```python
# disordered region
import numpy as np
import matplotlib.pyplot as plt
import random

def generate_random_number(min_value, max_value):
    return random.uniform(min_value, max_value)

Num_BN_in_one_column = 20
Num_La_positions = Num_BN_in_one_column + 1
La_positions_1_or_0 = np.zeros(Num_La_positions, dtype=int)

Num_columns_examined_for_disordered = 2000

p_have_La_disordered = 0.55

Zero_La_BN = [7.0, 10.5]
One_La_BN = [6, 10]
Two_La_BN = [6., 8]

BN_size_disordered = np.zeros(Num_columns_examined_for_disordered)
intensity_disordered = np.zeros(Num_columns_examined_for_disordered)

for i in range(Num_columns_examined_for_disordered):
    for k in range(Num_La_positions):
        p_La_choose = generate_random_number(0, 1)
        if p_La_choose > p_have_La_disordered:
            La_positions_1_or_0[k] = 0
        else:
            La_positions_1_or_0[k] = 1
    BN_size = 0
    for j in range(Num_BN_in_one_column):
        if La_positions_1_or_0[j] == 0 and La_positions_1_or_0[j+1] == 0:
            BN_size += generate_random_number(Zero_La_BN[0], Zero_La_BN[1])
        elif La_positions_1_or_0[j] == 1 and La_positions_1_or_0[j+1] == 1:
            BN_size += generate_random_number(Two_La_BN[0], Two_La_BN[1])
        else:
            BN_size += generate_random_number(One_La_BN[0], One_La_BN[1])
    intensity_La = np.sum(La_positions_1_or_0) * 57**2
    BN_size_disordered[i] = BN_size/Num_BN_in_one_column
    intensity_disordered[i] = intensity_La

max_intensity_possible = Num_La_positions * 57**2

fig = plt.figure()
ax = plt.axes()
normalized_intensity = intensity_disordered/max_intensity_possible + np.random.rand(intensity_disordered.shape[0])*.05
ax.scatter(normalized_intensity, BN_size_disordered, color='black')
plt.xlabel('Normalized La intensity')
plt.ylabel('Bottleneck area $(\AA^2)$')
plt.rcParams.update({'font.size': 14})
ax.set_xlim(0.15, .99)
ax.set_ylim(6.0, 9.9)
plt.show()
```

```python
# ordered region
import numpy as np
import matplotlib.pyplot as plt
import random

def generate_random_number(min_value, max_value):
    return random.uniform(min_value, max_value)

Num_BN_in_one_column = 20
Num_La_positions = Num_BN_in_one_column + 1
La_positions_1_or_0 = np.zeros(Num_La_positions, dtype=int)

Num_columns_examined_for_disordered = 2000

p_have_La_poor = 0.2
p_have_La_rich = 0.8

Zero_La_BN = [7.5, 10.5]
One_La_BN = [7.5, 10]
Two_La_BN = [6., 8]

BN_size_La_poor = np.zeros(Num_columns_examined_for_disordered)
intensity_La_poor = np.zeros(Num_columns_examined_for_disordered)

BN_size_La_rich = np.zeros(Num_columns_examined_for_disordered)
intensity_La_rich = np.zeros(Num_columns_examined_for_disordered)

# Let's calculate for the La-poor column
for i in range(Num_columns_examined_for_disordered):
    for k in range(Num_La_positions):
        p_La_choose = generate_random_number(0, 1)
        #print(p_La_choose)
        if p_La_choose > p_have_La_poor:
            La_positions_1_or_0[k] = 0
        else:
            La_positions_1_or_0[k] = 1
    BN_size = 0
    for j in range(Num_BN_in_one_column):
        if La_positions_1_or_0[j] == 0 and La_positions_1_or_0[j+1] == 0:
            BN_size += generate_random_number(Zero_La_BN[0], Zero_La_BN[1])
        elif La_positions_1_or_0[j] == 1 and La_positions_1_or_0[j+1] == 1:
            BN_size += generate_random_number(Two_La_BN[0], Two_La_BN[1])
        else:
            BN_size += generate_random_number(One_La_BN[0], One_La_BN[1])
    intensity_La = np.sum(La_positions_1_or_0) * 57**2
    BN_size_La_poor[i] = BN_size/Num_BN_in_one_column
    intensity_La_poor[i] = intensity_La
    
Zero_La_BN = [7.0, 10.5]
One_La_BN = [6, 10]
Two_La_BN = [6., 8]

# Let's calculate for the La-rich column
for i in range(Num_columns_examined_for_disordered):
    for k in range(Num_La_positions):
        p_La_choose = generate_random_number(0, 1)
        #print(p_La_choose)
        if p_La_choose > p_have_La_rich:
            La_positions_1_or_0[k] = 0
        else:
            La_positions_1_or_0[k] = 1
    BN_size = 0
    for j in range(Num_BN_in_one_column):
        if La_positions_1_or_0[j] == 0 and La_positions_1_or_0[j+1] == 0:
            BN_size += generate_random_number(Zero_La_BN[0], Zero_La_BN[1])
        elif La_positions_1_or_0[j] == 1 and La_positions_1_or_0[j+1] == 1:
            BN_size += generate_random_number(Two_La_BN[0], Two_La_BN[1])
        else:
            BN_size += generate_random_number(One_La_BN[0], One_La_BN[1])
    intensity_La = np.sum(La_positions_1_or_0) * 57**2
    BN_size_La_rich[i] = BN_size/Num_BN_in_one_column
    intensity_La_rich[i] = intensity_La
    
max_intensity_possible = Num_La_positions * 57**2

# Scattered plot for statistics
fig = plt.figure()
ax = plt.axes()

normalized_intensity_La_poor = intensity_La_poor/max_intensity_possible + np.random.rand(intensity_La_rich.shape[0])*.05
normalized_intensity_La_rich = intensity_La_rich/max_intensity_possible + np.random.rand(intensity_La_rich.shape[0])*.05

ax.scatter(normalized_intensity_La_poor, BN_size_La_poor, color='blue')
ax.scatter(normalized_intensity_La_rich, BN_size_La_rich, color='red')

plt.xlabel('Normalized La intensity')
plt.ylabel('Bottleneck area $(\AA^2)$')
plt.rcParams.update({'font.size': 14})
ax.set_ylim(6.0, 9.9)
plt.show()
```
